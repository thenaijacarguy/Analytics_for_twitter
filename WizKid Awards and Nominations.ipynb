# Next step, check to see if we have permission to import the table from WikiPedia

A 200 status code tells us that it's alright to access information from a certain website



```python
wikiurl='https://en.wikipedia.org/wiki/List_of_awards_and_nominations_received_by_Burna_Boy'
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

Now, we proceed to import the table using Beautiful Soup

```python
soup=BeautifulSoup(response.text, 'html.parser')
wiz = soup.find('table', {'class': 'wikitable'})
```

Assigning the table a name 'wizkid'

```python
wizkid = pd.read_html(str(wiz))
```

```python
wizkid
```

# Converting the table to a Pandas DataFrame


```python
wizkid = pd.DataFrame(wizkid[0])
```

Now, let's see what the table looks like after being converted to a dataframe



```python
wizkid
```

# Dropping unwanted columns

The last column is not necessary and needs to go

```python
wizkid=wizkid.drop('Ref', axis=1)
```

# Making a column the index
